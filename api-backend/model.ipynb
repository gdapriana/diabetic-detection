```python
import pandas as pd
import tensorflow as tf
import numpy as np
from numpy import genfromtxt
from sklearn.model_selection import train_test_split
import csv
```

```python
X_train = genfromtxt('dataset/x_train.csv', delimiter=',')
y_train = genfromtxt('dataset/y_train.csv', delimiter=',')
with open('dataset/headers.txt', newline='') as f:
    features = list(csv.reader(f))[0]
feature_len = len(features) - 1
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_train, y_train, test_size=0.25, random_state=0)
print('X_train:',np.shape(X_train))
print('y_train:',np.shape(y_train))
print('X_test:',np.shape(X_test))
print('y_test:',np.shape(y_test))
```

```python
model = tf.keras.Sequential([
  tf.keras.layers.Dense(units=16, activation='relu', input_shape=(feature_len, )),
  tf.keras.layers.Dense(1, activation='sigmoid')
])

adam = tf.keras.optimizers.Adam(learning_rate=0.01)
model.compile(loss='binary_crossentropy', optimizer=adam, metrics=["accuracy"])
```

```python
model.fit(X_train, y_train, epochs=300)
```

```python
loss_and_metrics = model.evaluate(X_test, y_test)
print(loss_and_metrics)
print('Loss = ',loss_and_metrics[0])
print('Accuracy = ',loss_and_metrics[1])
```
